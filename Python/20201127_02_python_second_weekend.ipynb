```python
def last_word(sentence):
    return sentence.split()[-1]

cuentito = 'Erase una vez'
last_word(cuentito)
```

```python
path = '/home/dsc/Data/shell/Finn.txt'
with open(path) as f:
    lines = f.readlines()

len(lines)
```

```python
def process_A(lines):
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(last_word(line))
    
    return result

process_A(lines)
```

```python
def process_B(lines):
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(line[0])
    
    return result

process_B(lines)
```

```python
def process(f, lines):
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(f(line))
    
    return result

process(len,lines)
```

### 2)

```python
names = ['Luis', 'Juan', 'Guille', 'Inés', 'Álvaro', 'Pablo', 'Miguel', 'Ana', 'María', 'Lucía']
```

```python
def toLowCase(x):
    return x.lower()

def upperEveryOtherLetter(x):
    str = ''
    for i in range(0,len(x)):
        if i%2 == 0 :
            newChar = x[i]
        else:
            newChar = x[i].upper()
    
        str += newChar
    return str        

def exercise2(names):
    namesLowerCase = list(map(toLowCase, names))
    return list(map(upperEveryOtherLetter, namesLowerCase))
    #namesEveryOtherLetterUpperCase = list(map(lambda x: ))

print(exercise2(names))
    
    
    
```
